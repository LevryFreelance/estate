```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup as Soup
```

```python
driver = webdriver.PhantomJS()
```

```python
html = Soup(requests.get('https://flatfy.lv/geo_list?geo_type=city&letter=3&section_id=1').text)
```

```python
for elem in html.select('.hscroll-selector__option.symbol-paging__item a'):
    print(elem['href'])
    link = 'https://flatfy.lv/' + elem['href']
    driver.get(link)
    print(driver.page_source)
    break
```
